```python
import pandas as pd
import sklearn
import warnings 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
df_house_train = pd.read_csv(r"/Users/hunglun/Downloads/train.csv")
df_house_test = pd.read_csv(r"/Users/hunglun/Downloads/test.csv")
```

```python
df_house_train.set_index('Id',inplace = True)
df_house_test.set_index('Id',inplace = True)
```

```python
df_SalePrice_train = df_house_train.pop('SalePrice')
```

```python
#Deleting Colums with half of the entries are NA
def DropNa(df):
    for (columnName, columnData) in df.iteritems():
        if df[columnName].isnull().sum() > len(df)//2:
            df.pop(columnName)
    return df
```

```python
def FillNa(df):
    for (columnName, columnData) in df.iteritems():
        if df[columnName].isnull().sum() != 0:
            if df[columnName].dtype == 'int64' or df[columnName].dtype == 'float64':
                df[columnName] = df[columnName].fillna(df[columnName].mean())
            else:
                df[columnName] = df[columnName].fillna(df[columnName].mode()[0])
    return df
        
```

```python
df_house_train = DropNa(df_house_train)
df_house_test = DropNa(df_house_test)
df_house_train = FillNa(df_house_train)
df_house_test = FillNa(df_house_test)
```

```python
df_final = pd.concat([df_house_train,df_house_test],axis = 0)
```

```python
df_house_train
```

```python
def CreateDummies(df):
    for (columnName, columnData) in df.iteritems():
        if df[columnName].dtype == 'object':
            df = pd.concat([df,pd.get_dummies(df[columnName])],axis = 1)
            df.pop(columnName)
    return df
        
```

```python
df_final = CreateDummies(df_final)
```

```python
df_final = df_final.loc[:,~df_final.columns.duplicated()]
```
